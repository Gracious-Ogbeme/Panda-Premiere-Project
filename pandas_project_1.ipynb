***Loading Data***

```python
# Importing libraries for analysis

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("Clean Dataset.xlsx")
```

```python
Years = data.columns[1:]
```

***Added New Column Called "Yearly Average***

```python
# Add yearly average as a new column to the dataframe

data['Avg GDP%'] = data.loc[:, Years].mean(axis=1)
```

***Performing Further EDA Using Chart***

```python
# Visualiazion of The Yearly Average GDP% by Country

plt.figure(figsize=(22, 6))
sns.barplot(data=data, x='Country Name', y='Avg GDP%', order=data.sort_values('Avg GDP%', ascending=False)['Country Name'])
plt.title(' Yearly Average GDP% by Country')
plt.ylabel('Country')
plt.xlabel('Yearly Average GDP%')
plt.xticks(rotation=45)
plt.show()
```

The chart represent the avarage yearly spending of 19 countries from the year 1999 to year 2021. The insight derived is that  Denmark, Norway and Sweden show a high yearly average GDP. While Japan, cote d'ivoire and Madagascar show relatively low GDP.

Correlation heatmap is very helpful when dealing wih the numerical feature/attribute. Here +1 (values in deep red) indicate a high positive correlation and -1(value in dark blue) indicate a negative correlation. From the chat above we can see that majority of the value are positively correlated. 

```python
#univarariate analysis on the distribution of average GDP across different countries

plt.figure(figsize=(10,6))
sns.histplot(data=data, x='Avg GDP%', kde=True, bins=5)
plt.title('Distribution of Average GDP ')
plt.xlabel('Average GDP')
plt.ylabel('Frequency')
plt.show()
```

From the chart above we can measure the distribution of the dataset. Here we can deduce that the distribution is positively skew, showing that majority of countries have a relatively low average GDP

```python
plt.figure(figsize=(20, 6))
for country in data['Country Name'].unique():
    country_data = data[data['Country Name'] == country]
    plt.plot(country_data.columns[1:-1], country_data.iloc[0, 1:-1], label=country)
plt.title('Education Spending Over Time')
plt.xlabel('Year')
plt.ylabel('Percentage of GDP Spent on Education')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
fig, axes = plt.subplots(3,3, figsize=(15, 9), sharex=True)
axes = axes.flatten()
plt.figure(figsize=(20, 6))
for country in data['Country Name'].unique():
    country_data = data[data['Country Name'] == country]
axes.plt(years, country_data.iloc[0, 1:-1].values.flatten(), marker='o', color='b', label=country)
axes.set_title('Education Spending Over Time')
axes.set_xlabel('Year')
axes.set_ylabel('Percentage of GDP Spent on Education')
axes.set_xticks(years)
axes.tick_params(axis='x', rotation=45)
axes.grid(True)
plt,show()
```
