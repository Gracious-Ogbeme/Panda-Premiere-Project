***Loading Data***

```python
# Importing libraries for analysis

import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Education_Spending_cleaned_data.csv")
```

***Exploratory Data Analysis***

```python
data.head()
```

***Added a new column called "Total Spending"***

```python
#Adding new column that sum the total spending from 2000 to 2021

Total_Spending = data.iloc[:, 4:].sum(axis=1)

# Add total spending as a new column to the dataframe

data['Total Spending'] = Total_Spending
data.head()
```

```python
# Showing the information about the data type

data.info()
```

```python
# Describing (overview) the numerical columns in our dataset 

data.describe()
```

```python
#Checking For Null values

data.isnull().sum()
```

***Performing Further EDA Using Chart***

```python
# Visualiazion of the total spending by country

plt.figure(figsize=(22, 6))
sns.barplot(data=data, x='Country Code', y='Total Spending')
plt.title('Total Educational Spending by Country')
plt.ylabel('country')
plt.xlabel('Total Spending')
plt.xticks( rotation=45)

plt.show()
```

```python
# Heatmap of spending across all years

plt.figure(figsize=(12,8))
sns.heatmap(data.iloc[:, 4:].corr(), cmap='coolwarm', annot=True)
plt.title('correlaton heatmap of educational spending across years')
plt.show()
```

```python
#univarariate analysis on the distribution of total spending across different countries

plt.figure(figsize=(10,6))
sns.histplot(data=data, x='Total Spending', kde=True)
plt.title('Distribution of Total Educational Spending ')
plt.xlabel('Total Spending')
plt.ylabel('Frequency')
plt.show()
```

```python
# we melt the dataframe tho convert the wide formate (with years a

melted_data = data.melt(id_vars=['Country Name', 'Country Code, 'Indicator Name'.
```

```python
# Quanity of they countries spending in education

edu_spending_level = ['Low', 'Medium', 'High']
def categorize_edu_spending_level(TotalSpending):
    if TotalSpending < 500:
        return 'Low'
    elif TotalSpending < 1000:
        return 'Medium'
    else:
        return 'High'

```

```python
# Plot Showing the comparison of average total spending in levels

data['edu_spending_level'] = data['Total Spending'].apply(categorize_edu_spending_level)

plt.figure(figsize=(10,6))
sns.barplot(data=data, x='edu_spending_level', y='Total Spending', order=edu_spending_level)
plt.title('Average Total Educational Spending by Level ')
plt.xlabel('education spending level')
plt.ylabel('Average Total Spending')
plt.show()
```

```python
plt.figure(figsize=(2,10))
sns.scatterplot(data=data, x='Total Spending', y='')
plt.title('Average Total Educational Spending by Level ')
plt.xlabel('education spending level')
plt.ylabel('Average Total Spending')
plt.show()
```
